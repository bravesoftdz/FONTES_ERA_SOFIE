```python
import csv
```

```python
def faixa_idade(idade: int):
    if 16 <= idade <= 20:
        return '1 - ADOLESCENTE'
    elif 21 <= idade <= 30:
        return '2 - JOVEM'
    elif 31 <= idade <= 40:
        return '3 - ADULTO'
    elif 41 <= idade <= 50:
        return '4 - MADURO'
    elif idade >= 51:
        return '5 - MELHOR IDADE'
    else:
        return 'DESCONHECIDO'
```

```python
with open(r'C:\fontes_SoFIE\backend_apps\analytics\OPINIONBOX_032019\original.csv', encoding='utf-8') as arquivo:
    arquivo.read(1)
    reader = csv.DictReader(arquivo, delimiter=';')
    
    print('Pessoa;Tarefa;Idade;Sexo;Classe')
    
    for each in reader:
        for job in each['P9'].split(';'):
            print(';'.join([each['Pessoa'], job.strip(), faixa_idade(int(each['Idade'])), each['Sexo'], each['Classe Social']]))          
```
