```python
import pandas as pd
from pytz import timezone

sp = timezone('America/Sao_Paulo')
```

```python
from pymongo import MongoClient
from pymongo.database import Database

database = MongoClient('mongodb://172.30.1.20:50102')

database['admin'].authenticate('worker', 'worker@sofie')

coll_sofier = database['config']['sofier']
coll_transaction = database['config']['transaction']
```

```python
def count_transactions(sofier):
    return coll_transaction.count_documents({'actors.sofier': sofier})

def count_sales(sofier, company):
    if sofier is None:
        return 0
    
    return coll_transaction.count_documents({
        'actors.sofier': sofier, 
        'actors.company': company,
        'status.status': 'FINISHED', 
        'status.success': True
    })
```

```python
rows = list()
for sofier in coll_sofier.find(projection={'_id': 0,'sofier_id': 1, 'email':1, 'indication_code': 1, 'name': 1, 'short_name': 1, '__created__.when': 1}):
    rows.append({
        'a_code': sofier.get('indication_code'),
        'a_email': sofier.get('email','').split('@')[0], 
        'b_short_name': sofier['short_name'],
        'c_data_cadastro': sofier['__created__']['when'], 
        'd_transactions': count_transactions(sofier.get('sofier_id')),
        'e_caras': count_sales(sofier.get('sofier_id'), 'caras'),
        'f_abril': count_sales(sofier.get('sofier_id'), 'abril'),
        'g_flor_de_sal': count_sales(sofier.get('sofier_id'), 'flor_de_sal'),
    })
    
planilha = pd.DataFrame(rows)

planilha
```
