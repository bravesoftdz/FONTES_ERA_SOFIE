```python
from pymongo import MongoClient
from pymongo.database import Database

database = MongoClient('mongodb://172.30.1.20:50102')

database['admin'].authenticate('worker', 'worker@sofie')

coll_sofier = database['config']['sofier']
coll_lead = database['config']['lead']
```

```python
for sofier in coll_sofier.find(projection={'__created__.when': 1, 'short_name': 1, 'email': 1, 'main_phone': 1}):
    print(f"{sofier['__created__']['when'].strftime('%d/%m/%Y')};{sofier['short_name']};{sofier.get('main_phone','')};{sofier['email']}")
```

```python
for item in coll_lead.find():
    print(f"{item['lead_name']};{item.get('phone','') or ''};{item.get('email','').lower()}")
```
