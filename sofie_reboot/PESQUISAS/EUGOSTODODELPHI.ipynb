```python
from requests import get, auth

CREDENTIALS = ('19678335832', 'spfc2010')

PROTOCOL = 'https'
HOST = 'mysofie.com'
PORT = 443
#: PROTOCOL = 'http'
#: HOST = 'localhost'
#: PORT = 64000


def build_item(item: dict):
    """
    Analisa um item gerando um texto HTML para ser postado no blog
    """
    comissao = 'R$ {:.2f}'.format(item['max_reward']).replace('.', ',')
    
    fragment = f"""
    <tr>
        <table>
            <td><img src="https://eugostododelphi.mysofie.com/api/v2/card/{item['name']}/cover" alt="{item['title']['title']}" height="120" width="100"></td>
            <td>
                <table>
                    <tr><td><strong>{item['title']['title']}</strong></td></tr> 
                    <tr><td>Modalidade: {item['title']['sub_title']}</td></tr> 
                    <tr><td>Período: {item['bookmarks']['card']['row_1']} {item['bookmarks']['card']['row_2']} - {item['bookmarks']['card']['row_3']}</td></tr> 
                    <tr><td><em>Comissão do blog: {comissao}</em></td></tr> 
                    <tr><td><a href="https://eugostododelphi.mysofie.com/offer/{item['name']}" target="_blank"><button>COMPRAR</button></a></td></tr> 
                </table>
            </td>
        </table>
    </tr>
    """
    return fragment
```

### RECUPERAÇÃO DOS CARDS DE VENDAS

Aqui recuperamos os cards de vendas e consideramos apenas os que estamos homologados

```python
URL = f'{PROTOCOL}://{HOST}:{PORT}/api/v2/sofier/me/challenge/cards/?expertise=SALE'
cards = get(URL, auth=CREDENTIALS)
```
